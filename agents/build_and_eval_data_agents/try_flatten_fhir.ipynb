```python
from fhirpower import flatten_fhir
import json
```

```python
# random sample of patient files
patient_file_paths = [
    './fhir/Aaron697_Brekke496_2fa15bc7-8866-461a-9000-f739e425860a.json',
    './fhir/Danyell947_Kautzer186_132ff57e-a51d-425e-af89-16d95db75d0f.json',
    './fhir/Rocio28_Zboncak558_736aec9a-fba2-4b15-ab38-09af9998c035.json'
]

flattened_json_lines = []
for file_paths in patient_file_paths:
    with open(file_paths) as f:
        source_json = json.load(f)
    flattened_json = flatten_fhir(source_json)
    print(flattened_json)
    flattened_json_lines.append(flattened_json)

print(flattened_json_lines[1])
```
